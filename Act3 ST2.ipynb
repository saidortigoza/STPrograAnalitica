```python
from scipy import stats
from scipy.stats import pearsonr

print(f"Coeficientes de correlacion en la tabla peliculas")

r, p = stats.pearsonr(df['type_code'], df['release_year'])
print(f"Correlación Pearson: r={r}, p-value={p}")

r, p = stats.spearmanr(df['type_code'], df['release_year'])
print(f"Correlación Spearman: r={r}, p-value={p}")

r, p = stats.kendalltau(df['type_code'], df['release_year'])
print(f"Correlación Pearson: r={r}, p-value={p}")
```
